### Fazendo as primeiras visualizações

```python
#impontando a bibliocata para leitura dos dados
import pandas as pd
```

```python
#Atribuindo os dados CVS para variavel 
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
#imprimindo as primeiras informações do CSV
df.head()
```

```python
#imprimindo as informações sobre o arquivo CSV
df.info()
```

```python
#convertendo o tipo do dados em DATETIME
df['data'] = pd.to_datetime(df['data'])
```

```python
#importando a biblioteca matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plotando data vs temperatura
plt.plot(df['data'], df['temperatura'])
```

```python
#melhorando o tamanho do nosso gráfico
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['temperatura'])
```
