## Aula 1

### Fazendo as primeiras visualizações

```python
#impontando a bibliocata para leitura dos dados
import pandas as pd
```

```python
#Atribuindo os dados CVS para variavel 
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
#imprimindo as primeiras informações do CSV
df.head()
```

```python
#imprimindo as informações sobre o arquivo CSV
df.info()
```

```python
#convertendo o tipo do dados em DATETIME
df['data'] = pd.to_datetime(df['data'])
```

```python
#importando a biblioteca matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plotando data vs temperatura
plt.plot(df['data'], df['temperatura'])
```

```python
#melhorando o tamanho do nosso gráfico
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['temperatura'])
```

```python
#adicionando o titulo
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
```

```python
#setando os parametros de um grafico em uma variavel
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
```

```python
#plotando o grafico dentro do eixo de imagem criado
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
```

```python
#identificação com legenda para cada eixo
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
eixo.set_ylabel('Temperatura')
eixo.set_xlabel('Data')
```

```python
#Alterando o tamanho da fonte
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
```

```python
#acrescentando legenda para o grafico
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
#acrescentando cor no gráfico
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

### Customizando nossas visualizações

```python
#aumentando a espessura da linhas
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', lw = 4)

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
#diminuido a espessura da linhas
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', lw = 0.4)

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
#alterando o estilo de linha
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', ls = 'dotted')

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
#importando a biblioteca de datas
import datetime
```

```python
#alterando a linha do tempo do gráfico
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
#adicionando marcadores no grafico
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', marker = 'o')

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
#adicionando grid
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
eixo.grid(True)
```
